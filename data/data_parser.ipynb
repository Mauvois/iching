# Création du Dataframe

```python
import pandas as pd
import re
```

```python
file_path = 'exported_text.txt'
# data = []
```

```python
with open(file_path, 'r', encoding='utf-8') as file:
    full_text = file.read()
```

```python
def parse_hexagram(hexagram):
    lines = hexagram.strip().split('\n')
    
    try:
        # Extract hexagram number and name
        number_name = lines[0].split('. ')
        number = int(number_name[0])
        name = number_name[1]

        # Extract top and bottom trigrams
        top_trigram = lines[1].split(', ')[-1]
        bottom_trigram = lines[2].split(', ')[-1]

        # Extract the judgment
        judgment_lines = []
        traits_index = None
        for i, line in enumerate(lines[3:], start=3):
            if line.startswith('Les traits'):
                traits_index = i
                break
            judgment_lines.append(line)
        judgment = ' '.join(judgment_lines).replace('\n', ' ')

        # Extract traits
        traits_lines = lines[traits_index+1:] if traits_index else []
        traits_text = '\n'.join(traits_lines)
        trait_pattern = re.compile(r'((Six|Neuf) [^\n]+\n(.*?)(?=\n(Six|Neuf) |$))', re.DOTALL)
        traits = trait_pattern.findall(traits_text)

        def clean_text(text):
            return text.replace('\n', ' ').strip()

        trait_1 = clean_text(traits[0][0]) if len(traits) > 0 else ''
        trait_2 = clean_text(traits[1][0]) if len(traits) > 1 else ''
        trait_3 = clean_text(traits[2][0]) if len(traits) > 2 else ''
        trait_4 = clean_text(traits[3][0]) if len(traits) > 3 else ''
        trait_5 = clean_text(traits[4][0]) if len(traits) > 4 else ''
        trait_6 = clean_text(traits[5][0]) if len(traits) > 5 else ''
        
        return {
            'Number': number,
            'Name': name,
            'Top Trigram': top_trigram,
            'Bottom Trigram': bottom_trigram,
            'Judgment': judgment,
            'Trait 1': trait_1,
            'Trait 2': trait_2,
            'Trait 3': trait_3,
            'Trait 4': trait_4,
            'Trait 5': trait_5,
            'Trait 6': trait_6,
        }
    except Exception as e:
        # Log the error and the problematic hexagram
        print(f"Error parsing hexagram: {lines[0]}")
        print(f"Error: {e}")
        return None

# Function to extract hexagrams from the full text
def extract_hexagrams(text):
    pattern = re.compile(r'(\d+\.\s[\s\S]+?)(?=\d+\.\s|$)')
    hexagrams = pattern.findall(text)
    return hexagrams

# Extract hexagrams from the full text
hexagrams = extract_hexagrams(full_text)

# Initialize empty list to store parsed hexagrams
parsed_hexagrams = []
```

```python
for hexagram in hexagrams:
    parsed_hexagram = parse_hexagram(hexagram)
    if parsed_hexagram:
        parsed_hexagrams.append(parsed_hexagram)

# Create DataFrame from parsed hexagrams
df = pd.DataFrame(parsed_hexagrams)
```

```python
df.to_csv("tabular_data", index=False)
```
